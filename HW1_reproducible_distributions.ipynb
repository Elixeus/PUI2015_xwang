# I want you to create a <i>reproducible research</i> code
that means we want the reader to be able to reproduce every step of our analysis. when stochastic processes and random variables are involved that gets tricky, but since computers are unable to rpoduce truly random numbers, we can actually reproduce every series of "pseudo random" number generated by a computer as long as we know what seed is used. <i>The first seed will determin entirely the distribution that is generated, and the seeds of the following distriburions generated by the code!</i>

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#THIS IS THE MOST IMPORTANT PIECE OF THIS CODE! 
##we must give numpy as seed for this code to be reproducible
http://docs.scipy.org/doc/numpy/reference/generated/numpy.random.seed.html

```python
seed = np.random.seed(10)
```

##use this function to generate 100 2-dimensional random datapoints in a gaussian distribution in 2 dimensions

http://docs.scipy.org/doc/numpy/reference/generated/numpy.random.randn.html

```python
n1 = 100
seed
distribution1 = np.random.randn(n1,2)
```

## use the same function to generate 50 more random objects, but this time change the mean and standard deviation of the gaussian distribution
to do that you multiply the distribution by the different sigma (one sigma, the same for each dimension), and add your new mean - remember that these are 2D objects, so yor mean must be a 2D vector (a list, an array or a tuple)! 


```python
n2 = 50
mean = [5, 5]
sigma = 2

seed
distribution2 = np.random.randn(n2,2)*sigma + mean
```

##now plot them with this function: 

https://scipy-lectures.github.io/intro/matplotlib/matplotlib.html#scatter-plots

##NOTE: if you print the array it will be a 2x100: 100 2D objects. although that is what we wanted we cannot plot it easily this way: there are many ways to deal with it. the goal is to pass the first dimension as the disrt argument to scatter, and the second dimension as the second argument. i am gonna do it by transposing the array (switching the axes) 

the command should look something like 

firstdistrib = firstdistrib.T 

which will replave your current variable firstdistrib, where you stored your first distribution, to the transpose of itself (i.e. switch columns and rows)

```python
x1 = distribution1[:,0]
y1 = distribution1[:,1]

x2 = distribution2[:,0]
y2 = distribution2[:,1]

plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.show()
```
